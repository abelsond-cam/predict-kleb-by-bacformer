# Analysis of Pangenome by Lineage

```python
import pandas as pd
import matplotlib.pyplot as plt

analysis_file = "/home/dca36/rds/rds-floto-bacterial-4k08a2yyQLw/david/processed/pangenome_analysis/gpa_reference_batch_summary_20260401_204443.tsv"

# Load the analysis file
analysis_df = pd.read_csv(analysis_file, sep="\t")

display(analysis_df.head())
```

```python
# print the list of columns as a list
print(analysis_df.columns.tolist())
```

```python
# Histogram of mean genome size between 4300 - 4800
plt.hist(analysis_df["mean_genome_size"], bins=range(4300, 4800, 50))
```

```python
# Plot this in 3 separate subplots using the same axis, varied by run_classification
fig, axs = plt.subplots(3, 1, sharey=True)
for i, run_classification in enumerate(analysis_df["run_classification"].unique()):
    axs[i].hist(analysis_df[analysis_df["run_classification"] == run_classification]["mean_genome_size"], bins=range(4300, 4800, 50))
    axs[i].set_title(run_classification)
plt.show()
```
